## 5. Price Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.distplot(avocado['AveragePrice'])
plt.title ("Distribution of Average Price")
```

```python
sns.barplot(x='AveragePrice',y='type',data=avocado) # organic avocado is more expensive than conventional one
plt.title ("Average price for each avocado type")
```

## 6. Avocado prices for different regions and years

### For Organic Avocado

```python
organicavocado = avocado[avocado['type'].isin(['organic'])]
organicavocado = organicavocado.sort_values(by = 'AveragePrice') 
#Although the highest avg.price of organic avocados was in 2016 in SanFrancisco
# in 2017 it has the higher average prices in most of the regions of USA.
```

```python
sns.catplot(x='AveragePrice', y='region',data=organicavocado ,hue='year', height=15, aspect=1.2, palette='Set1')
```

```python
sns.factorplot(x='AveragePrice', y='region',data=organicavocado ,hue='year', height=15, aspect=1.2, palette='Set1')
```

### For Conventional Avocado

```python
convenavocado = avocado[avocado['type'].isin(['conventional'])]
convenavocado = convenavocado.sort_values(by = 'AveragePrice') 
# in 2017 conventional avocados have the higher average prices in most of the regions of USA.
```

```python
sns.catplot(x='AveragePrice', y='region',data=convenavocado ,hue='year', height=15, aspect=1.2, palette='Set1')
```

```python
sns.factorplot(x='AveragePrice', y='region',data=convenavocado ,hue='year', height=15, aspect=1.2, palette='Set1')
```
